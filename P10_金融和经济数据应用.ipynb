### 频率不同的时间序列的运算

```python
gdp=pd.Series(np.abs(randn(7)),index=pd.period_range('1984Q2',periods=7,freq='Q-SEP'))
infl=pd.Series([0.025,0.045,0.037,0.04],index=pd.period_range('1982',periods=4,freq='A-DEC'))
```

```python
infl_q=infl.asfreq('Q-SEP',how='end')
infl_q
```
