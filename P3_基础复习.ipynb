### 异常处理

```python
def attempt_float(x):   
    try:
        return float(x)
    except:
        return x
attempt_float('something')      
```

```python
attempt_float((1,2))   #只要有错误就返回原值的做法
```

```python
def attempt_floatt(x):
    try:
        return float(x)
    except ValueError:
        return x
attempt_floatt('something')    #输入值不正确产生ValueError的做法
```

```python
attempt_floatt((1,2))
```

### 闭包：返回函数的函数

```python
def make_watcher():
    have_seen={}
    def has_been_seen(x):
        if x in have_seen:
            return True
        else:
            have_seen[x]=True
            return False
    return has_been_seen
watcher=make_watcher()
valss=[1,2,3,4,1,2,3,4]
[watcher(x)for x in valss]
```

```python
def make_change(amount,coins=[1,5,10,25],hand=None):
    hand=[] if hand is None else hand
    if amount==0:
        yield hand
    for coin in coins:
        if coin>amount or (len(hand)>0 and hand[-1]<coin):
            continue
        for result in make_change(amount-coin,coins=coins,hand=hand+[coin]):
            yield result      #迭代器的返回方式。
#amount=0时或者对每个coin上一步continue全执行时，就不迭代result，否则迭代result        
```

```python
for way in make_change(100,coins=[10,25,50]):
    print way    #输出结果的方式，最后输出的结果全是由hand产生，yield只起到了一个迭代的作用
```
